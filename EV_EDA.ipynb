## **What is Exploratory Data Analysis?**



*   Exploratory Data Analysis (EDA) is the process of analyzing a dataset in order
to understand its main characteristics, patterns and identify anomalies. EDA is
often the first step in the data analysis process.

*   It involves using different graphs and plots to help visualise the data and also uses statistical methods to draw inferences from the data.

*   The goal of EDA is not to arrive at a certain right answer or to confirm a pre-defined hypothesis. It is an exploratory process to draw inferences and get ideas on how the data can be further utilised to predict certain outcomes/develop ML models















# **Introduction to the Electric Vehicles**


*   An electric vehicle (EV) is a vehicle that uses one or more electric motors for propulsion. It can be powered by a collector system, with electricity from extravehicular sources, or it can be powered autonomously by a battery (sometimes charged by solar panels, or by converting fuel to electricity using fuel cells or a generator).

*   EVs include, but are not limited to, road and rail vehicles, surface and underwater vessels, electric aircraft , and electric spacecraft.
For road vehicles, together with other emerging automotive technologies such as autonomous driving, connected vehicles, and shared mobility, EVs form a future mobility vision called Connected, Autonomous, Shared, and Electric (CASE) Mobility.

*   EVs first came into existence in the late 19th century, when electricity was among the preferred methods for motor vehicle propulsion, providing a level of comfort and ease of operation that could not be achieved by the gasoline cars of the time. Internal combustion engines were the dominant propulsion method for cars and trucks for about 100 years, but electric power remained commonplace in other vehicle types, such as trains and smaller vehicles of all types.

# **Range - Battery Pack**

```python
#range-batterypack
def range_batterypack(dataframe, text):
    fig = plt.figure(figsize=(20,5))
    a1= plt.subplot()
    a1.bar(dataframe["CarName"], dataframe["Range_Km"], label='Range (Km)', color='green')
    plt.legend(loc= "upper left", bbox_to_anchor=(0,1.105))
    a2 = a1.twinx()
    a2.scatter(dataframe["CarName"], dataframe["Battery_Pack Kwh"], label= "Battery Pack", color= 'black')
    plt.title('''RANGE(Km) vs BATTERY PACK CAPACITY (kwh) of EV's costing {}'''.format(text), fontsize=12)
    a1.set_xlabel('Models', size= 16)
    a1.set_ylabel('Range (Km)', color = 'blue')
    a2.set_ylabel('Battery Pack Capacity (Kwh)', color='black')
    plt.legend(loc='upper left', bbox_to_anchor=(0,1))
    a1.set_xticklabels(df_1['CarName'], rotation = 'vertical')
    plt.show()

range_batterypack(df_1, t1)
range_batterypack(df_2, t2)
```

##**10. Compare Tesla, Volkswagen, Audi across multiple performance metrics?**

```python
# Selecting relevant columns for comparison
performance_metrics = ['AccelSec', 'TopSpeed_KmH', 'Range_Km', 'Efficiency_WhKm', 'FastCharge_KmH']
brands = ['Tesla', 'Volkswagen', 'Audi']

# Filter data for selected brands
filtered_df = df[df['Brand'].isin(brands)]

# Calculate average performance metrics for each brand
avg_performance = filtered_df.groupby('Brand')[performance_metrics].mean()

# Normalize the data for radar chart
min_vals = df[performance_metrics].min()
max_vals = df[performance_metrics].max()
normalized_data = (avg_performance - min_vals) / (max_vals - min_vals)

# Plotting radar chart
labels = np.array(performance_metrics)
stats = normalized_data.values

# Number of variables
num_vars = len(labels)

# Compute angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# The plot is circular, so we need to "complete the loop" and append the start data to the end.
stats = np.concatenate((stats, stats[:,[0]]), axis=1)
angles += angles[:1]

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Draw one axe per variable and add labels
plt.xticks(angles[:-1], labels)

# Plot data
for i in range(len(brands)):
    ax.plot(angles, stats[i], linewidth=2, linestyle='solid', label=brands[i])
    ax.fill(angles, stats[i], alpha=0.25)

# Adding legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Adding title
plt.title('Comparison of Tesla, Volkswagen, and Audi Across Performance Metrics')

# Show plot
plt.show()

```

Comparing Tesla, Volkswagen, and Audi across performance metrics, we can observe the following:
- Tesla generally excels in acceleration, top speed, and range, with its models having lower acceleration times, higher top speeds, and longer ranges compared to Volkswagen and Audi.
- Volkswagen vehicles tend to have relatively higher efficiency and fast charging speeds compared to Tesla and Audi, indicating efficient energy usage and faster charging capabilities.
- Audi strikes a balance between performance and efficiency, offering competitive performance metrics across acceleration, top speed, and range while maintaining good efficiency and fast charging capabilities.

```python
# Filter data for SUV body style electric vehicles
suv_data = df[df['BodyStyle'] == 'SUV']

# Calculate average acceleration for each powertrain type among SUVs
avg_acceleration_by_powertrain = suv_data.groupby('PowerTrain')['AccelSec'].mean().reset_index()

# Visualization: Bar plot comparing average acceleration among powertrain types for SUVs
plt.figure(figsize=(10, 6))
sbn.barplot(x='AccelSec', y='PowerTrain', data=avg_acceleration_by_powertrain, palette='Set2', hue='PowerTrain', dodge=False, legend=False)
plt.title('Average Acceleration among Powertrain Types for SUVs')
plt.xlabel('Average Acceleration (0-60 mph) in Seconds')
plt.ylabel('Powertrain Type')
plt.grid(True)
plt.show()

# Get the powertrain type with the highest average acceleration among SUVs
top_powertrain = avg_acceleration_by_powertrain.loc[avg_acceleration_by_powertrain['AccelSec'].idxmax()]

# Print the answer
print(f"The powertrain type with the highest average acceleration among SUVs is: {top_powertrain['PowerTrain']} with an average acceleration of {top_powertrain['AccelSec']:.2f} seconds.")
```

## **12. In which county tesla is famous?**

```python
# Filter the dataset for Tesla models
tesla_df = df[df['Brand'] == 'Tesla']

# Group by County and count the number of Tesla models in each county
tesla_count_by_county = tesla_df['County'].value_counts().reset_index()
tesla_count_by_county.columns = ['County', 'Tesla Count']

# Sort the counties by the number of Tesla models in descending order
tesla_count_by_county = tesla_count_by_county.sort_values(by='Tesla Count', ascending=False)

# Visualization: Doughnut chart showing the distribution of Tesla models across different counties
plt.figure(figsize=(12, 10))
explode = [0.1 if c == tesla_count_by_county.iloc[0]['County'] else 0 for c in tesla_count_by_county['County']]
plt.pie(tesla_count_by_county['Tesla Count'], labels=None, autopct='%1.1f%%', startangle=140, colors=sbn.color_palette('rainbow', len(tesla_count_by_county)), wedgeprops=dict(width=0.4), explode=explode)
plt.title('Distribution of Tesla Models Across Different Counties', fontsize=16)
plt.axis('equal')
plt.legend(tesla_count_by_county['County'], loc='best', title='County', fontsize='medium')
plt.show()

# Calculate and print the top percentage
top_county = tesla_count_by_county.iloc[0]  # Assuming the DataFrame is already sorted
total_models = tesla_count_by_county['Tesla Count'].sum()
top_percent = (top_county['Tesla Count'] / total_models) * 100
print(f"The county with the highest number of Tesla models ({top_county['County']}) has {top_percent:.1f}% of all Tesla models.")
```

we can apply the following criteria:

- Price less than $56,000
- High efficiency (lower Efficiency_WhKm values indicate higher efficiency)
- Large battery pack capacity
- Long range per charge
- Fast charging capability
